#### This program is written to extract the elevation (in meters) and the index locations of latitude and longitude for the study location from a TIFF file (e.g., created using GEO4PALM).

```python
import rasterio
from rasterio.warp import transform
from rasterio.enums import Resampling
```

```python
## Required files/inputs

# Path to your TIFF file (e.g., from GEO4PALM)
tif_file = "H:/Munich_campaign_analysis/PALM_4U/Jobs/1_pre_jobs/12km_3km_nest/Static_driver/Munich_static_dem_N02.tif"

# Latitude and Longitude of the study location (e.g., sensor location in 'EPSG:4326')  
lat_req = 48.14822338217075
lon_req = 11.573342795025054
```

```python
# Function to get the value at a specific lat/lon

def get_value_at_lat_lon(tif_file, lat_req, lon_req):

    with rasterio.open(tif_file) as dataset:

        # Convert latitude and longitude to the dataset's coordinate reference system (CRS)
        dst_crs = dataset.crs
        src_crs = 'EPSG:4326'
        lon, lat = transform(src_crs, dst_crs, [lon_req], [lat_req])

        # Get row and column corresponding to the transformed coordinates
        col, row = dataset.index(lon[0], lat[0])

        # Read the value at the specific location
        value = dataset.read(1, window=((col, col + 1), (row, row + 1)))
        return value[0][0], len(lat), (col, row)
```

```python
# Get the value at the specified location

value, len_lat, (col, row) = get_value_at_lat_lon(tif_file, lat_req, lon_req)
print(f"Elevation is {value} meters")
print(f"Corresponding longitude index(column), latitude index (row): {col}, {row}")
print(f"Corresponding longitude index(column), latitude index (row) in the PALM domain: {col}, {row-len_lat}")
```
